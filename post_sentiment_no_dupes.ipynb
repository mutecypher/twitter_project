```python
import numpy as np
import pandas as pd
```

```python
filez = ['/Volumes/Elements/GitHub/twitter-project/Data Files/CRWD_nn_scored.csv', 
'/Volumes/Elements/GitHub/twitter-project/Data Files/APPN_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/GMED_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/INSP_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/SNY_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/EVBG_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/KMI_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/Amazon_nn_scored.csv',
'/Volumes/Elements/GitHub/twitter-project/Data Files/Exxon_nn_scored.csv']


dataframes = []
i = 0
for filename in filez:
    print("i is", i)
    dataframes.append(pd.read_csv(filename, header='infer', index_col=0, parse_dates=True))
    i = i+1
Original_data = ['CRWD_df', 'APPN_df',  
          'GMED_df',
          'INSP_df', 'SNY_df','EVBG_df','KMI_df','Amzn_df', 'XOM_df']
Namez = ['CRWD_df', 'APPN_df',  
          'GMED_df',
          'INSP_df', 'SNY_df','EVBG_df','KMI_df','Amzn_df', 'XOM_df']
Keep_last = Original_data
Keep_first = Original_data

for i in range(len(Keep_last)):
    Original_data[i] = dataframes[i]
    ##Original_data[i].columns = ['number','created_at','full_text', 'retweet_count', 'user_id', 'user.favourites_count', 
    ##                 'user.followers_count','dupe','Symbol', 'Name', 'neg', 'neu', 'pos']
  ##  Original_data[i] = Original_data[i].sort_values('created_at', ascending = True)
    print("\nthe shape of ", Namez[i]," is ", Original_data[i].shape)
    Keep_last[i] = Original_data[i].drop_duplicates(subset = ['full_text'], keep='last', ignore_index = True)
    print("\nThe news shape of ", Namez[i], " is ", Keep_last[i].shape)
    
for i in range(len(Keep_first)):
    Original_data[i] = dataframes[i]
    ##Original_data[i].columns = ['number','created_at','full_text', 'retweet_count', 'user_id', 'user.favourites_count', 
    ##                 'user.followers_count','dupe','Symbol', 'Name', 'neg', 'neu', 'pos']
    Original_data[i] = Original_data[i].sort_values('neu', ascending = True)
    print("\nthe shape of ", Namez[i]," is ", Original_data[i].shape)
    Keep_first[i] = Original_data[i].drop_duplicates(subset = ['full_text'], keep='first', ignore_index = True)
    print("\nThe news shape of ", Namez[i], " is ", Keep_first[i].shape)
    
```
